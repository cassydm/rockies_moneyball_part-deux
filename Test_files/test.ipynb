```python
import pandas as pd
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.baseball-reference.com/teams/COL/2024.shtml#team_batting"
response= requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
tb = soup.find("table",id='team_batting')
df = pd.read_html(str(tb))[0]
df.head(21)
```

```python
df.plot(type="Bar")
```
