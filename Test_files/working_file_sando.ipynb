```python
# Function to convert salary to numerical value
def convert_salary(salary):
    if pd.isna(salary):
        return 0
    if isinstance(salary, str):
        salary = salary.replace('$', '').replace(',', '').replace('All-Star', '').strip()
        if salary.isnumeric():
            return int(salary)
        return 0
    return salary
```

```python
# Function to determine primary position
primary_positions = ['P', 'C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'OF', 'DH', 'PH', 'PR']

def determine_primary_position(row):
    for position in primary_positions:
        if position in row and row[position] > 0:
            return position
    return 'Unknown'
```

```python
# Function to calculate statistics for a given dataframe
def calculate_statistics(df):
    average_war = df['WAR'].mean()
    players_war_above_1 = df[df['WAR'] > 1.0]
    max_war_player = df.loc[df['WAR'].idxmax()]
    df['Salary'] = df['Salary'].apply(convert_salary)
    average_salary = df['Salary'].mean()
    df['Primary_Position'] = df.apply(determine_primary_position, axis=1)
    average_war_by_position = df.groupby('Primary_Position')['WAR'].mean()
    return average_war, players_war_above_1, max_war_player, average_salary, average_war_by_position
```

```python
average_war_2007, players_war_above_1_2007, max_war_player_2007, average_salary_2007, average_war_by_position_2007 = calculate_statistics(team_roster_2007)
average_war_2024, players_war_above_1_2024, max_war_player_2024, average_salary_2024, average_war_by_position_2024 = calculate_statistics(team_roster_2024)
```

```python
print(f"2007 - Average WAR: {average_war_2007}, Average Salary: ${average_salary_2007:,.2f}")
print(f"2024 - Average WAR: {average_war_2024}, Average Salary: ${average_salary_2024:,.2f}")
```

```python
#line chart comparison of average WAR score by postion
plt.figure(figsize=(12, 6))
plt.plot(average_war_by_position_2007.index, average_war_by_position_2007.values, label='2007', marker='o')
plt.plot(average_war_by_position_2024.index, average_war_by_position_2024.values, label='2024', marker='o')
plt.xlabel('Position')
plt.ylabel('Average WAR')
plt.title('Average WAR by Position (2007 vs 2024)')
plt.legend()
plt.grid(True)
plt.show()
```
