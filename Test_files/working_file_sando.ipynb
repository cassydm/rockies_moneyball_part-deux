```python
# Importing Dependencies
import pandas as pd
import requests
from bs4 import BeautifulSoup
import Links as ls
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
print(ls.Team_Roster_2007_url)
print(ls.Team_Roster_2024_url)
```

```python
roster_url = ls.Team_Roster_2024_url
response = requests.get(roster_url)
```

```python
team_roster_2024 = pd.read_excel ("../Test_files/Rockies_24_appearances.xlsx")
team_roster_2024
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(team_roster_2024['Age'], bins=10, kde=True)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/Age distribution 2024.png")
```

```python
team_roster_2024['Salary'] = team_roster_2024['Salary'].replace('[\$,]', '', regex=True).astype(float)

# Plotting the salary distribution of players
plt.figure(figsize=(10, 6))
sns.histplot(team_roster_2024['Salary'], bins=10, kde=True)
plt.title('Salary Distribution of Players')
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/Salary Distribution of Players 2024.png")
```

```python
# Plotting WAR vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=team_roster_2024, x='Salary', y='WAR')
plt.title('WAR vs. Salary')
plt.xlabel('Salary ($)')
plt.ylabel('WAR')
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/WAR vs Salary 2024.png")
```

```python
# Creating a bar chart for the number of games played by players in different positions
positions = ['C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'OF', 'DH', 'PH', 'PR']
games_by_position = team_roster_2024[positions].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
games_by_position.plot(kind='bar')
plt.title('Games Played by Position')
plt.xlabel('Position')
plt.ylabel('Total Games')
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/Games Played by Posistion 2024.png")
```
