```python
#Team roster for 2024
team_roster_2024 = pd.read_excel ("../Test_files/Rockies_24_appearances.xlsx")
team_roster_2024
```

```python
#Team roster for 2007
team_roster_2007 = pd.read_excel ("../Test_files/Rockies_07_appearances.xlsx")
team_roster_2007
```

```python
#roster for 2007 and 2024
team_roster_2024["Roster Year"]=2024
team_roster_2007["Roster Year"]=2007

team_roster_2024.head()
team_roster_2007.head()
```

```python
#concatinating roster for 2007 and 2024
combined_season_roster = pd.concat([team_roster_2007, team_roster_2024])
combined_season_roster.tail()

combined_season_roster.info()
```

```python
#scatter plot diagram of 2007 and 2024
plt.figure(figsize=(12, 6))
colors = {2007: ls.Rockies_Purple_Hex, 2024: ls.Rockies_Silver_Hex}
for season in combined_season_roster['Roster Year'].unique():
    subset = combined_season_roster[combined_season_roster['Roster Year'] == season]
    plt.scatter(subset['Salary'], subset['WAR'], c=colors[season], label=season, alpha=0.7)

plt.xlabel('Salary ($)')
plt.ylabel('WAR')
plt.title('WAR vs Salary (2007 vs 2024)')
plt.legend()
plt.grid(True)
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/scatter plot WAR vs Salary 2007 vs 2024.png")
```

```python
#vertical bar chart of 2007 and 2024
plt.figure(figsize=(14, 7))
colors = {2007: 'blue', 2024: 'green'}
for season in combined_season_roster['Roster Year'].unique():
    subset = combined_season_roster[combined_season_roster['Roster Year'] == season]
    plt.bar(subset['Name'], subset['Salary'], edgecolor=colors[season], label=season, alpha=0.5)

plt.xlabel('Name')
plt.ylabel('Salary ($)')
plt.title('Salary by Player (2007 vs 2024)')
plt.xticks(rotation=90)
# plt.legend()
plt.grid(True)
plt.tight_layout()
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/Salary by Player 2007 vs 2024.png")
```

```python
#combining the 2007 and 2024 roster salary and War score 
df_salary_war = combined_season_roster[['Salary', 'WAR']].dropna()
```

```python
# Calculate descriptive statistics
mean_salary = df_salary_war['Salary'].mean()
median_salary = df_salary_war['Salary'].median()
std_salary = df_salary_war['Salary'].std()

mean_war = df_salary_war['WAR'].mean()
median_war = df_salary_war['WAR'].median()
std_war = df_salary_war['WAR'].std()

print(f"Salary - Mean: {mean_salary}, Median: {median_salary}, Std Dev: {std_salary}")
print(f"WAR - Mean: {mean_war}, Median: {median_war}, Std Dev: {std_war}")
```

```python
from scipy.stats import pearsonr  
from sklearn.linear_model import LinearRegression
```

```python
# Calculate Pearson correlation coefficient
corr, _ = pearsonr(df_salary_war['Salary'], df_salary_war['WAR'])
print(f"Pearson correlation coefficient between Salary and WAR: {corr}")
```

```python
import seaborn as sns
```

```python
# Plot Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary', y='WAR', data=df_salary_war)
plt.title('Scatter Plot of Salary vs WAR')
plt.xlabel('Salary ($)')
plt.ylabel('WAR')
plt.grid(True)
plt.show()
```

```python
# Regression Analysis
X = df_salary_war[['Salary']]
y = df_salary_war['WAR']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
#scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary', y='WAR', data=df_salary_war, label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Salary vs WAR with Regression Line')
plt.xlabel('Salary ($)')
plt.ylabel('WAR')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Regression Coefficient: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Combine 2007 and 2024 data
df_combined = pd.concat([team_roster_2007.assign(Year=2007), team_roster_2024.assign(Year=2024)])
```

```python
# Drop rows with NaN values in Salary or WAR
df_clustering = df_combined[['Salary', 'WAR']].dropna()
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust n_clusters as needed
df_clustering['Cluster'] = kmeans.fit_predict(df_clustering[['Salary', 'WAR']])
```

```python
# Merge cluster labels with original data
df_combined = df_combined.merge(df_clustering[['Cluster']], left_index=True, right_index=True)
```

```python
# Visualize Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary', y='WAR', hue='Cluster', data=df_combined, palette='viridis', alpha=0.7)
plt.title('K-Means Clustering of Players Based on Salary and WAR')
plt.xlabel('Salary ($)')
plt.ylabel('WAR')
plt.legend(title='Cluster')
plt.grid(True)
#plt.show()
plt.savefig("../Test_files/Appearance_Figures/clustering of Players bases on salary and war 2007 vs 2024.png")
```

```python
# Ensure that 'Position' column is correctly created
position_columns = ['P', 'C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'OF', 'DH', 'PH', 'PR']
df_combined['Position'] = df_combined[position_columns].idxmax(axis=1)
```

```python
# Compute average Salary and WAR by Position and Year
position_salary_war = df_combined.groupby(['Year', 'Position'])[['Salary', 'WAR']].mean().unstack()
```
