```python
# Importing Dependencies
import pandas as pd
import requests
from bs4 import BeautifulSoup
import Links as ls
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
print(ls.Team_Roster_2007_url)
print(ls.Team_Roster_2024_url)
```

```python
team_roster_2024 = pd.read_excel ("../Test_files/Rockies_24_appearances.xlsx")
team_roster_2024
```

```python
team_roster_2007 = pd.read_excel ("../Test_files/Rockies_07_appearances.xlsx")
team_roster_2007
```

```python
team_roster_2024["Roster Year"]=2024
team_roster_2007["Roster Year"]=2007

team_roster_2024.head()
team_roster_2007.head()
```

```python
combined_season_roster = pd.concat([team_roster_2007, team_roster_2024])
combined_season_roster.tail()

combined_season_roster.info()
```

```python
plt.figure(figsize=(12, 6))
colors = {2007: 'blue', 2024: 'green'}
for season in combined_season_roster['Roster Year'].unique():
    subset = combined_season_roster[combined_season_roster['Roster Year'] == season]
    plt.scatter(subset['Salary'], subset['WAR'], c=colors[season], label=season, alpha=0.5)

plt.xlabel('Salary ($)')
plt.ylabel('WAR')
plt.title('WAR vs Salary (2007 vs 2024)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
colors = {2007: 'blue', 2024: 'green'}
for season in combined_season_roster['Roster Year'].unique():
    subset = combined_season_roster[combined_season_roster['Roster Year'] == season]
    plt.bar(subset['Name'], subset['Salary'], edgecolor=colors[season], label=season, alpha=0.5)

plt.xlabel('Name')
plt.ylabel('Salary ($)')
plt.title('Salary by Player (2007 vs 2024)')
plt.xticks(rotation=90)
# plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
