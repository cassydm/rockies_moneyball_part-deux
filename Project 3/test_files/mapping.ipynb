```python
import pandas as pd
```

```python
player_db = pd.read_csv('../data/Player_database_2024.csv')
player_db
```

```python
batter_db = pd.read_excel('../data/batter_mapping.xlsx')
batter_db = batter_db.drop(['Unnamed: 0'], axis= 1)
batter_db
```

```python
# Create a mapping dictionary from merged_df
embed_map = batter_db.set_index('Player_id')['Embedded_id'].to_dict()

# Map embed_id to player_db based on batter_id
player_db['Embedded_id'] = player_db['MLBAMID'].map(embed_map).astype('Int64')

# Display the updated player_db
player_db
```

```python
player_db.to_csv('../data/Player_database_2024_final.csv')
```
