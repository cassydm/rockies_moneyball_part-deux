```python
# import function
from pybaseball import statcast
from functions import team_stats 
from mega import mega_df_cleaned 
```

```python
# create copy dataframe
df_cleaned_strikeouts = mega_df_cleaned.copy()
df_cleaned_strikeouts.head()
```

```python

#apply stats function to the mega dataframe
df_cleaned_strikeouts['OPP_kk'] = df_cleaned_strikeouts.apply(lambda row: team_stats(row['Opp'], row['Formatted_Date'], "strikeouts"), axis=1)  
```

```python

# check 
df_cleaned_strikeouts.head()
```

```python

# export COL_hr column to separate csv file
COL_strikeouts_column = df_cleaned_strikeouts['OPP_kk']
COL_strikeouts_column.to_csv("progress_dataframes_opp/OPP_strikeouts_column.csv", header=True)
```
